# categorical graphs
 
  -columns that have txt are generally categorical
  -the first step is  to count the no. of unique values in column
  -then plot the count of each category
  -following charts a goods way to visualiz data

   -bar chart
   -pie chart
   -line chart
   -donut chart
   -area chart

  

```python
titanic['survived'].value_counts()
```

```python
survival_count = titanic['alive'].value_counts()
survival_count.plot(kind = 'pie', autopct = '%.1f%%', title = 'survival rate')
```

```python
survival_count.plot(
    kind = 'pie' ,
    autopct = '%.1f%%',
    title = 'survival rate',
    labels = ['survived' , 'not survived'],
    colors = ['#66b3ff' , '#ff9999'],
    labeldistance = 1.1,
    wedgeprops = {
        'width' : 0.5, # donut width
    }
)
```

```python
survival_count.plot(
    kind = 'pie' ,
    autopct = '%.1f%%',
    title = 'no. of ppl',
    labels = ['3' , '1' , '2'],
    colors = ['#66b3ff' , '#ff9999'],
    labeldistance = 1.1,
    wedgeprops = {
        'width' : 0.5, # donut width
    }
)
```
