```python
x = np.arange(-5 , 5 , .2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
fig , axs = plt.subplots(nrows = 1 , ncols = 3, figsize=(10 , 5))
axs[0].plot(x , y1, 'r--')
axs[1].plot(x , y2 , 'b--')
axs[2].plot(x , y3 , 'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
plt.show()
```

implicit vs explicit graphs


```python
# plt -> implicit figure and axes creation
x = np.linspace(-10 , 10 , 25)
plt.plot(x , x, '-')
plt.plot(x , x**2, '--' , label = 'quadratic' )
plt.plot(x , x**3 , 'o:', label = 'cubic')
plt.xlable('my x lable')
plt.ylable('my y lable')
plt.title('simple plot implicit graph setting ')
plt.legend()
plt.sjow
```

```python
# explicit
x = np.linspace(-10 , 10 , 25)
fig , ax = plt.subplots()
ax.plot(x , x, label = 'linear')
ax.plot(x , x**2, label = 'quadratic')
ax.plot(x , x**3, label = 'quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('simple plot')
ax.legend()
plt.show()
```

```python
data ={
    'name' : ['bob', 'jessica' , 'mary' , 'john' ,'mel'],
    'english': [70 , 40 , 60 , 80 , 90],
    'math' :   [80 , 86 , 90 , 40 , 30],
    'hindi' :  [71 , 50 , 30 , 59 , 60],

}
```

```python
def my_custom_plotter(ax , x_value , y_value , param_dict):
    out = ax.bar(x_value , y_value , **param_dict)
    return out
```

```python
fig , axs = plt.subplot(1 ,3 , figsize=(15 , 3))
my_custom_plotter(axs[0] , 'name' , 'english', {'data' : data , 'color' : 'red' , 'alpha': .3})
my_custom_plotter(axs[1] , 'name' , 'math', {'data' : data , 'color' : 'red' , 'alpha': .3})
my_custom_plotter(axs[2] , 'name' , 'hindi', {'data' : data , 'color' : 'red' , 'alpha': .3})
```
